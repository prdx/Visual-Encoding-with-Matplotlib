In this assignment, we want to compare using different encodings with the same goal. That is to visualize the average HP and Attack for each Pokemon type (Type 1).

```python
import pandas
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
```

```python
df = pd.read_csv("./Pokemon.csv")
df = df.dropna()
df = df.loc[:,["Type 1", "HP", "Attack"]]
```

--------------------------------------------------------------------------------------

```python
# Inspired from: https://idl.cs.washington.edu/files/2012-VisualizationZoo-CACM.pdf

avg = df.groupby(["Type 1"], as_index=False).mean()
plt.figure(figsize=(20,25))
plt.title(label="Statistical Distribution of Average HP and Attack for Each Pokemon Type", fontsize=20)
parallel_coordinates(avg, 'Type 1')
plt.legend(fontsize=20)

plt.show()
```

In this visualization, display one visualization to visualize statistical distributions
We encode the HP and attack, which both are quantitative as position channel with the line as the marker. The advantage of this visualization according to the original paper is this visualization is compact and so many variables can be shown simultanously while also allowing the reader to do pattern finding. Color is used as a channel to show each Pokemon type.

The disadvantage of this visualization, it is hard to see if we have so many datasets going to different directions. The solutions would be to sample the data from each category or show the average. That way the visualization looks less cluttered. Another disadvantage is that the reader would need training to read the plot [VAD]

```python
avg = df.groupby(["Type 1"]).mean()
avg.plot.bar(figsize=(40,15),fontsize=40,stacked=False)
plt.xlabel("Pokemon type", fontsize=40)
plt.legend(fontsize=30)
plt.title("Average attack and HP for each pokemon type\n", fontsize=40)
```

We now compare the first visualization with a simple bar chart. The reason we use bar chart because it is so effective in showing quantitative number for each categorical variable. We can see that it has high data-ink-ratio. 
Basically this visualization use the same marker (line) and position as a channel to show the quantitative value. The difference is only in the usage of the color channel, here we use color channel to differentiate between two categories HP and attack. We don't stack each category, as according to Weber’s Law, human judges based on relative not absolute difference, thus by grouping it makes the reader easy to get the relative difference.

### Between those two:
For pattern finding, it would be useful to use the parallel coordinates. However, for the bar plot is very effective in communicating the result yet very simple and easy to understand.

### Visualization encodings we don't use

In this assignment, we don't use area with point as mark and angle (tilt) and area as mark. Areas and angle channel are also can be used to describe quantitative data. However, using area as a channel together with point as mark will make one category overlapping to the other. While for tilt, since the category are too many (18 pokemon types), it would make the reader harder to read the chart and differentiate the angle. The example where we use area as mark with angle is pie-chart.
